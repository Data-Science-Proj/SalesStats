```python
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/BestBuyAPIs/open-data-set/master/products.json'
```

```python
response = requests.get(url)
```

```python
response_json = response.json()
```

```python
data = pd.DataFrame(response_json)
```

```python
data.head(30)
```

```python
data.shape
```

```python
data.columns
```

```python
data = data.drop(['category', 'image', 'model', 'shipping', 'sku', 'upc', 'url'], axis = 1)
```

```python
data.info()
```

```python
data.describe()
```

```python
np.unique(data.type)
```

```python
data.manufacturer=data.manufacturer.fillna("")
```

```python
np.unique(data.manufacturer)
```

```python
types = ['BlackTie', 'Game', 'HardGood', 'Movie', 'Music', 'Software']
def type_price_mean():
    avg_type_price=[0,0,0,0,0,0]
    for i in range(len(types)):
        mask = data['type'] == types[i]
        type_data = data.loc[mask].mean()
        avg_type_price[i]=type_data
    return avg_type_price

print(type_price_mean())
```

The data set is very large and it takes lots of time to process it, so I decided to work with a random sample of it.


```python
data_random = data.sample(n=15000)
```

```python
np.unique(data_random.type)
```

```python
types = ['BlackTie', 'Game', 'HardGood', 'Movie', 'Music', 'Software']
def type_price_mean():
    avg_type_price=[0,0,0,0,0,0]
    for i in range(len(types)):
        mask = data_random['type'] == types[i]
        type_data = data_random.loc[mask].mean()
        avg_type_price[i]=type_data
    return avg_type_price

print(type_price_mean())
```

```python
plt.figure(figsize=(8,5))
sns.barplot(types,type_price_mean(), palette="Blues_d");
plt.title('Average Price of Different Departments', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.xlabel('Department', fontsize=14)
plt.show()
```
